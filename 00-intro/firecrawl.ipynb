```python
from firecrawl import Firecrawl

firecrawl = Firecrawl(api_key="")

# Scrape a website:
doc = firecrawl.scrape("https://sesac.seoul.kr/course/active/detail.do", formats=["markdown", "html"])
print(doc)
```

```python
from pprint import pprint

with open('./my-crawl.md', 'w', encoding='utf-8') as f:
    f.write(doc.markdown)

pprint(type(doc.markdown))
```
